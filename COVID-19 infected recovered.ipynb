# Healed vs. Confirmed vs. Dead

```python
import pandas as pd
df = pd.read_csv('data/03-04-2020.csv')
df = df.drop(columns=['Last Update', 'Latitude', 'Longitude'])

overall = df.agg(['sum'])
```

```python
overall['infected'] = overall['Confirmed'] - overall['Recovered']
```

## Finally plot a simple bar chart

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))

ax = plt.gca()
plt.xlabel('Confirmed vs. Infected vs. Dead')
plt.ylabel('People')

# ugly, but it works...
plt.bar(['Confirmed', 'Recovered','Infected', 'Dead'], [overall['Confirmed'][0],overall['Recovered'][0], overall['infected'][0],overall['Deaths'][0]], color=['purple', 'green', 'blue', 'red'])
plt.title('Overall Confirmed vs. Infected vs. Recovered vs. Dead')

plt.savefig('healed_vs_confirmed_vs_infected', dpi=300)

plt.show()

```
